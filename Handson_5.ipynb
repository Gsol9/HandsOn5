```python
import random

MAX_ITEMS = 4

def read_items(filename):
    with open(filename, "r") as file:
        return [line.strip() for line in file if line.strip()]

def write_items(filename, items):
    with open(filename, "w") as file:
        for item in items:
            file.write(item + "\n")

def show_items(inventory):
    if len(inventory) == 0:
        print("Your inventory is empty.")
    else:
        for i, item in enumerate(inventory, start=1):
            print(f"{i}.\t{item}")

def walk(inventory, all_items, inventory_file):
    item = random.choice(all_items)
    print(f"While walking down a path, you see {item}.")
    choice = input("Do you want to grab it? (y/n): ").lower()

    if choice == "y":
        if len(inventory) >= MAX_ITEMS:
            print("You can't carry any more items. Drop something first.")
        elif item in inventory:
            print("You already have that item.")
        else:
            inventory.append(item)
            write_items(inventory_file, inventory)
            print(f"You picked up {item}.")
    else:
        print("You leave the item behind.")

def drop_item(inventory, inventory_file):
    if len(inventory) == 0:
        print("Nothing to drop.")
        return

    try:
        number = int(input("Number: "))
        if 1 <= number <= len(inventory):
            item = inventory.pop(number - 1)
            write_items(inventory_file, inventory)
            print(f"You dropped {item}.")
        else:
            print("Invalid item number.")
    except ValueError:
        print("Invalid number.")

def main():
    all_items_file = "wizard_all_items.txt"
    inventory_file = "wizard_inventory.txt"

    all_items = read_items(all_items_file)
    inventory = read_items(inventory_file)

    print("The Wizard Inventory Program")
    print()
    print("COMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program")

    while True:
        print()
        command = input("Command: ").lower()

        if command == "walk":
            walk(inventory, all_items, inventory_file)
        elif command == "show":
            show_items(inventory)
        elif command == "drop":
            drop_item(inventory, inventory_file)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
```

```python
import csv

def read_sales(filename):
    sales = []
    with open(filename, "r", newline="") as file:
        reader = csv.reader(file)
        for row in reader:
            if row:
                sales.append([row[0], float(row[1])])
    return sales

def write_sales(filename, sales):
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)
        for month, amount in sales:
            writer.writerow([month, int(amount) if amount == int(amount) else amount])

def show_monthly_sales(sales):
    for month, amount in sales:
        if amount == int(amount):
            amount = int(amount)
        print(f"{month} - {amount}")

def show_yearly_summary(sales):
    total = sum(amount for _, amount in sales)
    average = round(total / len(sales), 2)
    print(f"Yearly total:\t{int(total) if total == int(total) else total}")
    print(f"Monthly average:\t{average}")

def edit_sales(sales, filename):
    month = input("Three-letter Month: ").title()
    valid_months = [m for m, _ in sales]

    if month not in valid_months:
        print("Invalid three-letter month.")
        return

    try:
        amount = float(input("Sales Amount: "))
        for i in range(len(sales)):
            if sales[i][0] == month:
                sales[i][1] = amount
                write_sales(filename, sales)
                print(f"Sales amount for {month} was modified.")
                break
    except ValueError:
        print("Invalid sales amount.")

def main():
    filename = "monthly_sales.csv"
    sales = read_sales(filename)

    print("Monthly Sales program")
    print()
    print("COMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly - View yearly summary")
    print("edit - Edit sales for a month")
    print("exit - Exit program")

    while True:
        print()
        command = input("Command: ").lower()

        if command == "monthly":
            show_monthly_sales(sales)
        elif command == "yearly":
            show_yearly_summary(sales)
        elif command == "edit":
            edit_sales(sales, filename)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
```
